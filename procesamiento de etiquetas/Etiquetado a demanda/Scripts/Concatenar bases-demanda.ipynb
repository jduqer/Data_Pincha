# Cargar bases

```python
# merging the files
files_joined = os.path.join('..\\Datos partidos', "*.csv")

# Return a list of all joined files
list_files = glob.glob(files_joined)

#print("** Merging multiple csv files into a single pandas dataframe **")
# Merge files by joining all files
dataframe = pd.concat(map(pd.read_csv, list_files), ignore_index=True)
print(dataframe)
```

```python
columnas = ['Event','Tag Description','Categoria','Tipo','Tag Start','Tag End','Jugador','Zonas1','Zonas2']

dataframe = dataframe[columnas].reset_index(drop=True)
dataframe.columns = ['Partido','Situacion','Evento','Subevento','Tiempo_inicial','Tiempo_final','Jugador','Zonas_Inicial','Zonas_Final']
```

## Procesar tiempo

```python
'''TIEMPO'''

from datetime import timedelta


def seconder(x):
    mins, secs = map(float, x.split(':'))
    td = timedelta(minutes=mins, seconds=secs)
    return td.total_seconds()

dataframe['segundos'] = dataframe['Minuto en el video'].apply(seconder)
```
